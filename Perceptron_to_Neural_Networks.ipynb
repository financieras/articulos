# Perceptron to Neural Networks: The Journey Begins
## Una aproximación a las Redes Neuronales para programadores

## El Dataset
- Usaremos el **Dataset de Cáncer de Mama de Wisconsin** como hilo conductor.
- Es perfecto: binario (benigno/maligno), intuitivo y de interés general.
- Construiremos desde la neurona más simple hasta una red neuronal básica, enfatizando la analogía con la programación clásica.

## Índice
1. Introducción y "El Problema": El Dataset de Wisconsin
    - Objetivo: Enganchar. Contextualizar.
    - ¿Qué son los datos? (Características numéricas de células -> Diagnóstico).
    - Analogía para programadores: Es como tener una función `diagnosticar(características)` que no sabemos escribir con `if-else` porque la lógica es demasiado compleja y sutil. ¿Cómo la "creamos" automáticamente a partir de datos?
    - Presentación visual de los datos.
2. La neurona fundamental: El Perceptrón
    - Objetivo: Entender la unidad básica de computación.
    - Analogía: Un *nodo decisor* muy simple.
    - Explicación intuitiva con matemáticas accesibles:
        * Entradas (x) -> Pesos (w) -> Suma ponderada
        * El Bias (sesgo). Explicarlo como el "umbral" o el "nivel de exigencia" del perceptrón. Es el parámetro que determina cuándo se activa.
        * Función de Activación (Step Function): "Si la suma total supera el umbral, ENCIENDE (1), si no, APAGA (0)". Es como un `if (suma > umbral) return 1; else return 0;`.
    - Visual: Dibujo de un perceptrón con entradas, pesos, bias y salida.
3. ¿Cómo Aprende un Perceptrón? El "Ajuste de Tornillos" Automático
    - Objetivo: Lo principal es entender el *aprendizaje*.
    - Partimos de un perceptrón con **pesos y bias aleatorios** -> Da respuestas aleatorias.
    - **Función de Coste** (Error): Es una simple **medida de cuántas veces se equivoca**. Cuantos más errores, mayor el "coste".
    - **Descenso del Gradiente**:
        * Imaginemos un gráfico con los valores de los **pesos** en un eje y el **error** en otro. Es un "valle".
        * Nuestro perceptrón aleatorio está en una ladera de ese valle, con error alto.
        * Queremos bajar al fondo (error mínimo). **El gradiente es la pendiente de la ladera** donde estamos parados. Nos indica la dirección más pronunciada **hacia abajo**.
        * **Learning Rate** (Tasa de Aprendizaje): ¿De qué tamaño es el paso que damos? ¡No demasiado grande (podemos saltarnos el mínimo) ni demasiado pequeño (llegariamos muy lentamente)
    - Analogía: Es como ajustar los tornillos (pesos y bias) de una máquina viendo si su rendimiento mejora o empeora, pero con una brújula matemática (el gradiente) que nos dice exactamente qué tornillo ajustar y en qué dirección.
4. De una Neurona a un Cerebro: **La Red Neuronal**
    - Objetivo: Entender el poder de la combinación de neuronas.
    - Conectamos muchos perceptrones en capas, ahora con funciones de activación más suaves como **ReLU** o **Sigmoide**.
    - **Front Propagation**: Los datos entran por la izquierda, se transforman capa a capa y sale una predicción por la derecha. Es solo cálculo secuencial.
    - **Back Propagation**: Cuando la red se equivoca, el "culpable" no es solo la última neurona. El error se **propaga hacia atrás** para saber cuánto contribuyó al error cada neurona en cada capa y así ajustar **todos** sus pesos. Es el "reparto de culpas" algorítmico.
    - Visual: Red neuronal de 4 capas (entrada, 2 ocultas, salida) mostrando el flujo de datos (hacia adelante) y el flujo del error (hacia atrás).
5. Poniéndolo Todo en Práctica: Train, Test y Evaluación
    - **Separación Train/Test**: No podemos evaluar con los mismos datos con los que aprendimos (sería hacer trampa). Train (80%) para aprender, Test (20%) para ver si realmente generaliza.
    - Evaluación: **Accuracy** (Precisión) - ¿Qué porcentaje de casos del test acertó?
    - **Matriz de Confusión**: No es lo mismo equivocarse diagnosticando un cáncer benigno como maligno (falso positivo) que al revés (falso negativo, MUY GRAVE). La matriz nos da este detalle.
    - **Normalización**: Poner todas las características en la misma escala para que ninguna domine el aprendizaje por tener números más grandes. Por ejemplo, Z-score, haciendo que todas las características (features) tengan media cero y desviación típica 1.
    - **Early Stopping**: Dejar de entrenar cuando el error deja de mejorar.
    - **Overfitting** y **Underfitting**: Procurar evitar el sobreajuste y el infraajuste.
6. Conclusión
    - Recapitulación del viaje: De una neurona simple que aprende, a una red que aprende patrones complejos.
    - Resumen de los conceptos clave: Perceptrón, Pesos/Bias, Función de Activación, Aprendizaje = Ajuste de Pesos, Descenso del Gradiente, Red = Muchas Neuronas, Backpropagation, Evaluación Honesta (Train/Test).
